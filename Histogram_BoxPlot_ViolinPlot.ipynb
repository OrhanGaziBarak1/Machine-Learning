```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini yükle
df = pd.read_csv("newDataset.csv")

# Oluşturmak istediğiniz sütun isimleri
target_columns = ['Artist Popularity', 'Track Duration(ms)', 'Track Popularity', 'Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valance', 'Tempo', 'Time_signature']

# Her sütun için grafik oluştur
for column in target_columns:
    if column in df.columns and df[column].dtype != 'object':
        plt.figure(figsize=(10, 5))
        
        # Histogram
        plt.subplot(1, 3, 1)
        sns.histplot(df[column], kde=True)
        plt.title('Histogram of ' + column)
        plt.xlabel(column)
        plt.ylabel('Frequency')
        
        # Box Plot
        plt.subplot(1, 3, 2)
        sns.boxplot(y=df[column])
        plt.title('Box Plot of ' + column)
        plt.ylabel(column)
        
        # Violin Plot
        plt.subplot(1, 3, 3)
        sns.violinplot(y=df[column])
        plt.title('Violin Plot of ' + column)
        plt.ylabel(column)
        
        # Izgara ekle
        sns.set_style("whitegrid")
        
        plt.tight_layout()
        plt.show()
    else:
        print(f"'{column}'Column not found or does not contain numeric data.")
```
