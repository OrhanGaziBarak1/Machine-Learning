```python
##Mp3 klasöründeki ses dosyalarının her birinin spektrogram grafiği oluşturup bir dosyaya "png" şeklinde kaydeden kod
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

def create_output_folder(output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

def plot_spectrogram(mp3_file, output_dir):
    # Ses dosyasını yükle
    y, sr = librosa.load(mp3_file)
    
    # Spektrogramu oluştur
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    
    # Grafik çizimi
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spektrogram')
    plt.savefig(os.path.join(output_dir, os.path.splitext(os.path.basename(mp3_file))[0] + '.png'))
    plt.close()

def process_mp3_folder(folder_path, output_dir):
    create_output_folder(output_dir)  # Çıkış klasörünü oluştur
    # Klasördeki tüm dosyaları gez
    for filename in os.listdir(folder_path):
        if filename.endswith('.mp3'):
            mp3_file = os.path.join(folder_path, filename)
            plot_spectrogram(mp3_file, output_dir)

# Klasörler ve dosya yolları
input_folder = 'mp3_tracks'  # MP3 dosyalarının olduğu klasörün adı
output_folder = 'spektrogramlar'  # Spektrogramların kaydedileceği adı

# Çıkartma işlemi
process_mp3_folder(input_folder, output_folder)
```
