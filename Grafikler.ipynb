```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
veri = pd.read_csv("final_dataset.csv")
```

```python
veri.rename(columns = {"Artist Popularity" : "Artist_Popularity",
                       "Artist Genres": "Artist_Genres",
                       "Release Date" : "Release_Date",
                       "Track Popularity" : "Track_Popularity"}, inplace= True)
veri.head()
```

```python
plt.subplot(131)
plt.title('Plot Histogram')
plt.hist(veri.Artist_Popularity, color = "g")

plt.subplot(132)
plt.title('Box-Plot')
sns.boxplot(veri.Artist_Popularity,palette="Set3")

plt.subplot(133)
plt.title('Violin Plot')
sns.violinplot(veri.Artist_Popularity, linewidth=5, palette="Set2")
plt.show()
plt.tight_layout() 
```
