```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# final_dataset.csv dosyasını DataFrame'e yükleyin
df = pd.read_csv("final_dataset.csv")

# Sayısal sütunları seçin
numerical_columns = ["Artist Popularity", "Track Duration(ms)", "Track Popularity", 
                     "Danceability", "Energy", "Key", "Loudness", "Mode", 
                     "Speechiness", "Acousticness", "Instrumentalness", 
                     "Liveness", "Valence", "Tempo"]

# Her bir sütunu diğer sütunlarla işleme sokun
for output_column in numerical_columns:
    plt.figure(figsize=(10, 6))
    
    # Diğer sütunları gezerek işlem yapın
    for input_column in numerical_columns:
        # Aynı sütunlar arasında işlem yapmayın
        if input_column == output_column:
            continue
        
        # Scatter plot çizimi
        plt.scatter(df[input_column], df[output_column], alpha=0.5, label=input_column)
    
    # Grafik başlığı ve etiketleri
    plt.title(f"{output_column} vs Diğer Sütunlar")
    plt.xlabel("Diğer Sütunlar")
    plt.ylabel(output_column)
    plt.legend()
    
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# final_dataset.csv dosyasını DataFrame'e yükleyin
df = pd.read_csv("final_dataset.csv")

# Sayısal sütunları seçin
numerical_columns = ["Artist Popularity", "Track Duration(ms)", "Track Popularity", 
                     "Danceability", "Energy", "Key", "Loudness", "Mode", 
                     "Speechiness", "Acousticness", "Instrumentalness", 
                     "Liveness", "Valence", "Tempo"]

# Her bir sütun için ayrı grafikler çizin
for output_column in numerical_columns:
    plt.figure(figsize=(15, 10))
    
    # Diğer sütunları gezerek işlem yapın
    for i, input_column in enumerate(numerical_columns):
        # Aynı sütunlar arasında işlem yapmayın
        if input_column == output_column:
            continue
        
        plt.subplot(4, 4, i + 1)
        # Scatter plot çizimi
        plt.scatter(df[input_column], df[output_column], alpha=0.5)
        plt.title(f"{output_column} vs {input_column}")
        plt.xlabel(input_column)
        plt.ylabel(output_column)
    
    plt.tight_layout()
    plt.show()
```
