    Her nümerik input ve output sütunu için standardizasyon işlemi yapıldı.

```python
import pandas
from sklearn.preprocessing import StandardScaler

dataset = pandas.read_csv("playlist_dataset_operational.csv")

scaler = StandardScaler()
input_columns_standardized = scaler.fit_transform(dataset[["Artist Popularity", "Track Duration(ms)", "Track Popularity", 
                         "Danceability", "Energy", "Key", "Loudness", "Mode", "Speechiness",
                        "Acousticness", "Instrumentalness", "Liveness", "Valence", "Tempo", "Time_signature"]])
output_columns_standardized = scaler.fit_transform(dataset[["Artist Popularity", "Track Popularity"]])

data_standardized = pandas.DataFrame(data=input_columns_standardized, columns=["Artist Popularity", "Track Duration(ms)", "Track Popularity", 
                         "Danceability", "Energy", "Key", "Loudness", "Mode", "Speechiness",
                        "Acousticness", "Instrumentalness", "Liveness", "Valence", "Tempo", "Time_signature"])
data_standardized[["Artist Popularity Output", "Track Popularity Output"]] = pandas.DataFrame(data=output_columns_standardized,
                                                                                              columns=["Artist Popularity", "Track Popularity"])

data_standardized
#data_standardized.to_csv("data_standardized_for_numerical_data.csv")
```

    Her nümerik input ve output için normalleştirilmiş mutlak çapraz korelasyon haritası işlemi yapıldı. (Yüksek ihtimalle hatalı.)

```python
import pandas
import seaborn
import matplotlib.pyplot as plot

dataset = pandas.read_csv("playlist_dataset_operational.csv")

inputs = dataset.select_dtypes(include=['float64', 'int64'])
outputs = dataset[["Artist Popularity", "Track Popularity"]]

correlation_matrix = inputs.corrwith(outputs)

plot.figure(figsize=(12, 8))
seaborn.heatmap(correlation_matrix.abs().values.reshape(1, -1), cmap='viridis', annot=True, xticklabels=correlation_matrix.index, yticklabels=False)
plot.title('Normalized Absolute Cross-Correlation Map')
plot.xlabel('Variables')
plot.ylabel('Correlation with Outputs')
plot.show()
```
