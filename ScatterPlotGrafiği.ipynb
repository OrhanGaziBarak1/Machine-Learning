```python
import pandas as pd
data = pd.read_csv("C:\\Users\\Neslihan\\Desktop\\SayısalnewdatasetLyricli.csv")
print(data.head(100))
silinecek_sutunlar = ['Lyrics']

# Belirtilen sütunları DataFrame'den kaldırın
#data.drop(columns=silinecek_sutunlar, inplace=True)
data
```

```python
import matplotlib.pyplot as plt


# Özellik listesi
features = ["Track Duration(ms)", "Track Popularity", "Danceability", "Energy", "Loudness", 
            "Speechiness", "Acousticness", "Instrumentalness", "Liveness", "Valence", 
            "Tempo", "Time_signature"]

# Her bir özellik için scatter plot çizin
for feature in features:
    plt.figure(figsize=(8, 6))
    plt.scatter(data[feature], data["Artist Popularity"])
    plt.xlabel(feature)
    plt.ylabel("Artist Popularity")
    plt.title("{} vs Artist Popularity Scatter Plot".format(feature))
    plt.grid(True)
    plt.show()
```
