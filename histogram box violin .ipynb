```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# final_dataset.csv dosyasını DataFrame'e yükleyin
df = pd.read_csv("final_dataset.csv")

# Sadece sayısal sütunları seçin
numeric_cols = df.select_dtypes(include=[float, int]).columns

# Her bir sayısal sütun için histogram, kutu grafiği (box plot) ve keman grafiği (violin plot) çizin
for column in numeric_cols:
    plt.figure(figsize=(12, 6))
    
    # Histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f"{column} Histogram")
    
    # Kutu Grafiği (Box Plot)
    plt.subplot(1, 3, 2)
    sns.boxplot(y=df[column])
    plt.title(f"{column} Box Plot")
    
    # Keman Grafiği (Violin Plot)
    plt.subplot(1, 3, 3)
    sns.violinplot(y=df[column])
    plt.title(f"{column} Violin Plot")
    
    plt.tight_layout()
    plt.show()
```
