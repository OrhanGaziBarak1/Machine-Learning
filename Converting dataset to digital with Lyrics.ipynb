```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Veri setini yükle
data = pd.read_csv("new_dataset.csv")

# Silinecek sütunları belirtin
silinecek_sutunlar = ['Unnamed: 0']

# Belirtilen sütunları DataFrame'den kaldırın
data.drop(columns=silinecek_sutunlar, inplace=True)


# Lyrics sütununu ayrı bir değişkende sakla
lyrics_column = data['Lyrics']

# Lyrics sütununu veri setinden çıkar
data = data.drop(columns=['Lyrics'])

# Yalnızca string formatındaki sütunları seç
string_columns = data.select_dtypes(include=['object']).columns

# Label Encoding işlemi
label_encoder = LabelEncoder()
for column in string_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Lyrics sütununu yeniden DataFrame'e ekle
data['Lyrics'] = lyrics_column

print("Dönüştürülmüş Veri Seti:")
# Dönüştürülmüş veri setini CSV olarak kaydet
data.to_csv("SayısalnewdatasetLyricli.csv", index=False)

#print(data)
data
```
