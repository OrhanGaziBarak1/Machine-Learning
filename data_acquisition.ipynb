```python
import csv
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError

# YouTube API anahtarınızı buraya girin
API_KEY = 'key'

# YouTube API'sini yapılandır
youtube = build('youtube', 'v3', developerKey=API_KEY)

# CSV dosyasından şarkı isimlerini ve URL'leri almak için bir fonksiyon
def get_youtube_urls(input_csv_file):
    youtube_urls = []

    with open(input_csv_file, 'r', newline='', encoding='utf-8') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            song_name = row['Track Name']
            artist_name = row['Artist']
            youtube_url = get_youtube_url(song_name, artist_name)
            youtube_urls.append((song_name, artist_name, youtube_url))

    return youtube_urls

# YouTube araması yaparak şarkı ismine göre URL almak için bir fonksiyon
def get_youtube_url(song_name, artist_name):
    try:
        # YouTube API'sini kullanarak arama yap
        search_query = f"{song_name} {artist_name}"  # Şarkı ve şarkıcı adını birleştir
        request = youtube.search().list(
            part='snippet',
            q=search_query,
            type='video',
            maxResults=1
        )
        response = request.execute()

        # İlk sonucu kontrol et ve URL'yi döndür
        if 'items' in response:
            video_id = response['items'][0]['id']['videoId']
            youtube_url = f'https://www.youtube.com/watch?v={video_id}'
            return youtube_url
        else:
            return None

    except HttpError as e:
        print(f'YouTube API Error: {e}')
        return None

# ilk_10_satir_track_name.csv dosyasından şarkı isimlerine göre YouTube URL'lerini al
input_csv_file = 'ilk_10_satir.csv'
youtube_urls = get_youtube_urls(input_csv_file)

# Alınan URL'leri ekrana yazdır
for song_name, artist_name, youtube_url in youtube_urls:
    print(f"{song_name} - {artist_name}: {youtube_url}")
```
