```python
from collections import Counter
import string
import nltk
from nltk.corpus import words
import numpy as np

# İngilizce kelimeler listesi
nltk.download('words')
english_words = set(words.words())

# İngilizce'deki yaygın bağlaçlar ve edatlar listesi
baglac_edatlar = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]
df = pd.read_csv("final_dataset.csv")
# Her bir satırdaki metinleri noktalama işaretlerinden arındırarak kelime kelime ayırın
tum_kelimeler = []
for metin in df['Lyrics']:
    # Noktalama işaretlerini kaldır ve metni küçük harfe çevir
    metin = metin.translate(str.maketrans('', '', string.punctuation)).lower()
    kelimeler_metin = metin.split()
    # Bağlaçlar, edatlar ve ingilizce olmayan kelimeleri hariç tutarak kelimeleri ekle
    tum_kelimeler.extend([kelime for kelime in kelimeler_metin if kelime not in baglac_edatlar and kelime in english_words])

# Kelimelerin sayısını hesaplayın
kelime_sayilari = Counter(tum_kelimeler)

# En çok tekrar eden kelimenin sayısını ve en az tekrar eden kelimenin sayısını bulun
en_cok_tekrar_sayisi = max(kelime_sayilari.values())
en_az_tekrar_sayisi = min(kelime_sayilari.values())

# Normalizasyon için değer aralıklarını belirleyin
min_deger = 0
max_deger = 1

# Her kelimenin tekrar sayısını normalize edin
kelime_degerleri = {}
for kelime, sayi in kelime_sayilari.items():
    if sayi == en_cok_tekrar_sayisi:
        kelime_degerleri[kelime] = max_deger
    elif sayi == en_az_tekrar_sayisi:
        kelime_degerleri[kelime] = min_deger
    else:
        # Normalizasyon formülü: (değer - min) / (max - min)
        kelime_degerleri[kelime] = (sayi - en_az_tekrar_sayisi) / (en_cok_tekrar_sayisi - en_az_tekrar_sayisi)

# Sonuçları büyükten küçüğe sıralayın
kelime_degerleri_sirali = dict(sorted(kelime_degerleri.items(), key=lambda x: x[1], reverse=True))

# Sonuçları yazdırın
print("Kelime Değerleri (Büyükten Küçüğe Sıralı):")
for kelime, deger in kelime_degerleri_sirali.items():
    print(f"{kelime}: {deger}")
    
# Değerlerin kaç tane olduğunu bulun
deger_sayilari = Counter(kelime_degerleri.values())

# Sonuçları yazdırın
print("Değerlerin Sayısı:")
for deger, sayi in deger_sayilari.items():
    print(f"{deger}: {sayi} adet")
```
